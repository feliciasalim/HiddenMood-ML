```python
import pandas as pd
from googleapiclient.discovery import build
import time
```

## Scrape Youtube

```python
API_KEY = 'AIzaSyCClcNfzbBhqt-p4BDhLsbOoKEll1E2dkM'
youtube = build('youtube', 'v3', developerKey=API_KEY)

KEYWORDS = [
    "Anxious",
    "Depressed",
    "Overwhelmed",
    "Panic Attack",
    "Loneliness"
]

RESULTS_PER_KEYWORD = 50
results = []

for keyword in KEYWORDS:
    print(f"Searching for: {keyword}")
    request = youtube.search().list(
        q=keyword,
        part='snippet',
        type='video',
        maxResults=RESULTS_PER_KEYWORD
    )
    response = request.execute()

    for item in response.get('items', []):
        video_data = {
            'Keyword': keyword,
            'Title': item['snippet']['title'],
            'Link': f"https://www.youtube.com/watch?v={item['id']['videoId']}"
        }
        results.append(video_data)

    time.sleep(1)

df = pd.DataFrame(results)
# df.to_csv("youtube_keywords_dataset4.csv", index=False)
# print(f"Done! Saved {len(df)} rows to 'youtube_keywords_dataset.csv'")
```
